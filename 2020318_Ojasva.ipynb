PART C

```python
#c part
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import LabelEncoder
import numpy as np
data = pd.read_csv('CO2 Emissions.csv')
print(data.head())
```

```python
# Columns to be used for linear regression
reduced = data.copy()

# Label encoding for categorical columns
le = {}
for item in reduced.columns:
    if reduced[item].dtype == "object":
        le[item] = LabelEncoder()
        reduced[item] = le[item].fit_transform(reduced[item])

# Train:Test split 80:20
x = reduced.drop("CO2 Emissions(g/km)", axis=1)
y = reduced["CO2 Emissions(g/km)"]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)

```

```python
#Linear regression
model = LinearRegression()
model.fit(xtrain, ytrain)

#Prediction on train and test data
ytrain_predicted = model.predict(xtrain)
ytest_predicted = model.predict(xtest)

#MSE
mse_train = mean_squared_error(ytrain, ytrain_predicted)
mse_test = mean_squared_error(ytest, ytest_predicted)

#RMSE
rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

#R2 score
r2_train = r2_score(ytrain, ytrain_predicted)
r2_test = r2_score(ytest, ytest_predicted)

#R2 adjusted score for test data
n = len(ytest)
p = xtest.shape[1]
r2_adj_test = 1-((1 - r2_test)*(n - 1)/(n - p - 1))
r2_adj_train = 1-((1 - r2_train)*(n - 1)/(n - p - 1))

#MAE
maetrain = mean_absolute_error(ytrain, ytrain_predicted)
maetest = mean_absolute_error(ytest, ytest_predicted)

```

```python
# Print the vals
print("Train MSE:", mse_train, end=", ")
print("Test MSE:", mse_test)
print()
print("Train RMSE:", rmse_train, end=", ")
print("Test RMSE:", rmse_test)
print()
print("Train R2 Score:", r2_train, end=", ")
print("Test R2 Score:", r2_test)
print("Adjusted R2 Score (Train):", r2_adj_train, end=",")
print("Adjusted R2 Score (Test):", r2_adj_test)
print()
print("MAE on train:", maetrain, end=", ")
print("MAE on test:", maetest)
```

```python
#g part
from sklearn.linear_model import Lasso, Ridge

data = pd.read_csv('CO2 Emissions.csv')

# Columns to be used for linear regression
reduced = data.copy()

# Label encoding for categorical columns
le = {}
for item in reduced.columns:
    if reduced[item].dtype == "object":
        le[item] = LabelEncoder()
        reduced[item] = le[item].fit_transform(reduced[item])

# Train:Test split 80:20
x = reduced.drop("CO2 Emissions(g/km)", axis=1)
y = reduced["CO2 Emissions(g/km)"]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
#Lasso Model
model_lasso = Lasso(alpha=0.01)
model_lasso.fit(xtrain, ytrain)
#Prediction on train and test data
ytrain_predicted_lasso = model_lasso.predict(xtrain)
ytest_predicted_lasso = model_lasso.predict(xtest)
#MSE
mse_train_lasso = mean_squared_error(ytrain, ytrain_predicted_lasso)
mse_test_lasso = mean_squared_error(ytest, ytest_predicted_lasso)
#RMSE
rmse_train_lasso = np.sqrt(mse_train_lasso)
rmse_test_lasso = np.sqrt(mse_test_lasso)
#R2 score
r2_train_lasso = r2_score(ytrain, ytrain_predicted_lasso)
r2_test_lasso = r2_score(ytest, ytest_predicted_lasso)
#R2 adjusted score for test data
n = len(ytest)
p = xtest.shape[1]
r2_adj_lasso_test = 1-((1 - r2_test_lasso)*(n - 1)/(n - p - 1))
r2_adj_lasso_train = 1-((1 - r2_train_lasso)*(n - 1)/(n - p - 1))
#MAE
maetrain_lasso = mean_absolute_error(ytrain, ytrain_predicted_lasso)
maetest_lasso = mean_absolute_error(ytest, ytest_predicted_lasso)

# Print the vals
print("Train MSE:", mse_train_lasso, end=", ")
print("Test MSE:", mse_test_lasso)
print()
print("Train RMSE:", rmse_train_lasso, end=", ")
print("Test RMSE:", rmse_test_lasso)
print()
print("Train R2 Score:", r2_train_lasso, end=", ")
print("Test R2 Score:", r2_test_lasso)
print("Adjusted R2 Score (Train):", r2_adj_lasso_train, end=",")
print("Adjusted R2 Score (Test):", r2_adj_lasso_test)
print()
print("MAE on train:", maetrain_lasso, end=", ")
print("MAE on test:", maetest_lasso)
```
